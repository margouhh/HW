```python
M = int(input("Введите количество строк M: "))
N = int(input("Введите количество столбцов N: "))
K = int(input("Введите номер строки K (1 <= K <= M): "))
if K < 1 or K > M:
    print("Некорректный номер строки K.")
else:
    matrix = []
    print("Введите элементы матрицы:")
    for i in range(M):
        row = list(map(int, input().split()))
        matrix.append(row)
    sum_K = sum(matrix[K - 1])
    for elem in matrix[K - 1]:
        product_K *= elem
    print(f"Сумма элементов {K}-й строки:", sum_K)
    print(f"Произведение элементов {K}-й строки:", product_K)
```

```python
M = int(input("M:"))
N = int(input("N:"))
matrix = []
print("Введите элементы матрицы:")
for i in range(M):
    row = list(map(int, input().split()))
    if len(row) != N:
        print(f"Ошибка: строка {i + 1} должна содержать {N} элементов.")
        exit()
    matrix.append(row)
found = False
for j in range(N):
    column_sum = sum(matrix[i][j] for i in range(M))
    column_avg = column_sum / M
    if all(matrix[i][j] == column_avg for i in range(M)):
        print(f"Все элементы столбца {j + 1} равны среднему арифметическому: {column_avg}")
        found = True
        break
if not found:
    print("Столбец, в котором все элементы равны среднему арифметическому, не найден.")

```
