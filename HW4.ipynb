6. Описать процедуру DigitCountSum (K,C,S), находящую количество C цифр целого положительного числа K, а также их сумму S (K-входной, C и S- выходные параметры целого типа). С помощью этой процедуры найти количество и сумму цифр для кажого из пяти данных целых чисел.

```python
def DigitCountSum(K):
    C = 0
    S = 0
    while K > 0:
        digit = K % 10
        S += digit
        C += 1
        K //= 10
    return C, S
numbers = []
for i in range(5):
    num = int(input(f"Введите целое положительное число {i+1}: "))
    numbers.append(num)
for num in numbers:
    C, S = DigitCountSum(num)
    print(f"Число: {num}, Количество цифр: {C}, Сумма цифр: {S}")

```

С помощью этой процедуры упорядочить по возрастанию два данных набора из трёх чисел: image.png image.png image.png

```python
def SortInc3(A, B, C):
    numbers = [A, B, C]
    sorted_numbers = sorted(numbers)
    return sorted_numbers, numbers
A = float(input("A:"))
B = float(input("B:"))
C = float(input("C:"))
sorted_numbers, original_numbers = SortInc3(A, B, C)
print(sorted_numbers)
print(original_numbers)
```

```python
def is_prime(num):
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True
def goldbach_hypothesis(n):
    if n <= 2 or n % 2 != 0:
        return "Число должно быть четным и больше 2."
    primes = [i for i in range(2, n) if is_prime(i)]
    for p in primes:
        if is_prime(n - p):
            return f"{n} = {p} + {n - p}"
    return "Гипотеза Гольдбаха не подтверждена для данного числа."
n = int(input("Введите четное число n (n > 2): "))
result = goldbach_hypothesis(n)
print(result)
```

```python
def is_prime(num):
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True
def next_prime(n):
    if n <= 1:
        return 2
    prime = n
    while True:
        prime += 1
        if is_prime(prime):
            return prime
n = int(input(":"))
next_prime_number = next_prime(n)
print(f"Следующее простое число после {n}: {next_prime_number}")
```

```python
def multiply_large_numbers(num1, num2):
    digits1 = list(num1)
    digits2 = list(num2)
    result = [0] * (len(digits1) + len(digits2))
    for i in range(len(digits1) - 1, -1, -1):
        carry = 0
        for j in range(len(digits2) - 1, -1, -1):
            product = int(digits1[i]) * int(digits2[j]) + carry
            result[i + j + 1] += product % 10
            carry = product // 10
        result[i + len(digits2)] += carry
    while len(result) > 1 and result[0] == 0:
        result.pop(0)
    return ''.join(map(str, result))
num1 = input(":")
num2 = input(":")
result = multiply_large_numbers(num1, num2)
print(result)
```
